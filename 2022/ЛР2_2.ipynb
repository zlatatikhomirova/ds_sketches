```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
```

1. Сформируйте список (массив) элементов. С клавиатуры запрашиваем количество элементов в массиве, затем вводим сами значения. Найдите максимальное, минимальное значения и их место, а также среднее арифметическое введенных чисел.


```python
arr = []
n = int(input("Введите количество элементов в массиве "))
for _ in range(n):
    arr.append(float(input()))
i_max, i_min, max_num, min_num = -1, -1, float("-inf"), float(" inf")
summ = 0
for i, e in enumerate(arr):
    if e > max_num:
        max_num = e
        i_max = i
    if e < min_num:
        min_num = e
        i_min = i
    summ += e

mean = summ/n

print(f"Максимальное значение = {max_num}, его индекс в массиве = {i_max}")
print(f"Минимальное значение = {min_num}, его индекс в массиве = {i_min}")
print(f"Среднее арифметическое введённых чисел = {mean}")
```

4. Постройте различные графики и гистограмму.

```python
ax = axes3d.Axes3D(plt.figure())
i = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(i, i)
Z = X**2 + Y**2
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
plt.show()
```

```python
def Data():
        x = np.arange(-20, 20, 0.1)
        y = np.arange(-20, 20, 0.1)
        xgrid, ygrid = np.meshgrid(x, y)
        zgrid = np.cos(xgrid)*np.sin(ygrid)/(xgrid+ygrid)
        return xgrid, ygrid, zgrid

x, y, z = Data()

fig = pylab.figure()
axes = Axes3D(fig)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap=cm.jet)
pylab.show()
```

```python
def Data():
        x = np.arange(-20, 20, 0.1)
        y = np.arange(-20, 20, 0.1)
        xgrid, ygrid = np.meshgrid(x, y)
        zgrid = np.cos(xgrid)*np.cos(ygrid)
        return xgrid, ygrid, zgrid

x, y, z = Data()

fig = pylab.figure()
axes = Axes3D(fig)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap=cm.jet)
pylab.show()
```
