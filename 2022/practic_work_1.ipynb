```python
import pandas as pd
import numpy as np
```

```python
d = {a: [] for a in ["Пол", "Возраст", "Слишком низкая температура", "Оптимальная температура", "Слишком высокая температура"]}


def gen_data(n=50):

    # МОЛОДЫЕ ЖЕНЩИНЫ

    for i in range(15):
        low_temp = np.random.randint(low=17, high=22)
        high_temp = np.random.randint(low=24, high=32)
        optim_temp = (low_temp + high_temp)//2
        if optim_temp == high_temp:
            high_temp += 1
        if optim_temp == low_temp:
            low_temp -= 1
        age = np.random.randint(low=15, high=35)
        d["Пол"].append("женский")
        d["Возраст"].append(age)
        d["Оптимальная температура"].append(optim_temp)
        d["Слишком низкая температура"].append(low_temp)
        d["Слишком высокая температура"].append(high_temp)

    # ПОЖИЛЫЕ ЖЕНЩИНЫ

    for i in range(5):
        low_temp = np.random.randint(low=17, high=22)
        high_temp = np.random.randint(low=24, high=28)
        optim_temp = (low_temp + high_temp)//2
        if optim_temp == high_temp:
            high_temp += 1
        if optim_temp == low_temp:
            low_temp -= 1
        age = np.random.randint(low=62, high=75)
        d["Пол"].append("женский")
        d["Возраст"].append(age)
        d["Оптимальная температура"].append(optim_temp)
        d["Слишком низкая температура"].append(low_temp)
        d["Слишком высокая температура"].append(high_temp)

    # МОЛОДЫЕ МУЖЧИНЫ

    for i in range(20):
        low_temp = np.random.randint(low=12, high=20)
        high_temp = np.random.randint(low=23, high=29)
        optim_temp = (low_temp + high_temp)//2
        if optim_temp == high_temp:
            high_temp += 1
        if optim_temp == low_temp:
            low_temp -= 1
        age = np.random.randint(low=15, high=35)
        d["Пол"].append("мужской")
        d["Возраст"].append(age)
        d["Оптимальная температура"].append(optim_temp)
        d["Слишком низкая температура"].append(low_temp)
        d["Слишком высокая температура"].append(high_temp)

    # ПОЖИЛЫЕ МУЖЧИНЫ

    for i in range(5):
        low_temp = np.random.randint(low=17, high=20)
        high_temp = np.random.randint(low=24, high=28)
        optim_temp = (low_temp + high_temp)//2
        if optim_temp == high_temp:
            high_temp += 1
        if optim_temp == low_temp:
            low_temp -= 1
        age = np.random.randint(low=62, high=75)
        d["Пол"].append("мужской")
        d["Возраст"].append(age)
        d["Оптимальная температура"].append(optim_temp)
        d["Слишком низкая температура"].append(low_temp)
        d["Слишком высокая температура"].append(high_temp)

    # ДОШКОЛЬНИКИ

    for i in range(5):
        low_temp = np.random.randint(low=18, high=20)
        high_temp = np.random.randint(low=25, high=30)
        optim_temp = (low_temp + high_temp)//2
        if optim_temp == high_temp:
            high_temp += 1
        if optim_temp == low_temp:
            low_temp -= 1
        age = np.random.randint(low=7, high=11)
        gender = ("женский", "мужской")
        idx = np.random.randint(2)
        d["Пол"].append(gender[idx])
        d["Возраст"].append(age)
        d["Оптимальная температура"].append(optim_temp)
        d["Слишком низкая температура"].append(low_temp)
        d["Слишком высокая температура"].append(high_temp)

```

```python
gen_data()
```

```python
df = pd.DataFrame(d)
df.index = df.index + 1
df
```
