```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = []
with open('/content/drive/MyDrive/nto_2022/data1.csv', 'r') as f:
  for i in f:
    arr.append(int(f.readline().strip()[:-1]))
```

```python
def get_spectr(data):
    y = np.array(data.split(','), dtype=float)
    xf = np.fft.rfftfreq(y.size, 1./250)
    yf = np.abs(np.fft.rfft(y))
    idxs = np.where((8 <= xf) & (xf <= 13))[0]
    spectrum = sum(yf[idxs[0] :idxs[-1] + 1])
    return spectrum
```
