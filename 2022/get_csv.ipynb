```python
import os
import pandas as pd
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
cur_dir = "/content/drive/MyDrive/hkt"
```

```python
folders = ['animals_here',
            'broken',
            'broken_night',
            'broken_snow',
            'broken_summer',
            'no_animals_here']
```

```python
cnt = 0
for folder in folders:
    os.chdir(f"{cur_dir}/{folder}")
    files = os.listdir()
    print(len(files), f"{cur_dir}/{folder}", sep='\t')
    cnt += len(files)
print(f"\n{cnt}\tTotal")
```

```python
df = {'filename':[], 'broken':[], 'empty': [], 'animal':[]}
for folder in folders:
    os.chdir(f"{cur_dir}/{folder}")
    for f in os.listdir():
        df['filename'].append(f"{cur_dir}/{folder}/{f}")
        if 'broken' in folder:
            df['broken'].append(1)
            df['empty'].append(0)
            df['animal'].append(0)
        elif 'no_animals_here' == folder:
            df['broken'].append(0)
            df['empty'].append(1)
            df['animal'].append(0)
        elif 'animals_here' == folder:
            df['broken'].append(0)
            df['empty'].append(0)
            df['animal'].append(1)
#os.chdir('..')
```

```python
print(*map(len, [df['filename'], df['animal'], df['broken'], df['empty']]))
```

```python
df = pd.DataFrame(df).drop_duplicates()
df.tail(20)
```

```python
df.to_csv('/content/drive/MyDrive/hkt/df30092023.csv')
```
