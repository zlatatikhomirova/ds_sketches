```python
s1 = ">" + 10*"1" + 20*"2" + 30*"3"
s2 = ">" + 20*"2" + 30*"3" + 10*"1"
s3 = ">" + 30*"3" + 10*"1" + 20*"2"
def f(s):
    while ">1" in s or ">2" in s or ">3" in s:
        if ">1" in s: s = s.replace(">1", "22>")
        if ">2" in s: s = s.replace(">2", "2>")
        if ">3" in s: s = s.replace(">3", "1>")
    print(s, s.count("1"), s.count("2"), f"ans = {s.count('1')+ 2*s.count('2')}")

for s in (s1, s2, s3):
    f(s)
```
