Множество и его элементы
Говоря слово "множество", мы будем под этим понимать совокупность объектов произвольного рода, рассматриваемую как единое целое. Вместе с термином множество будут употребляться и его синонимы типа набор, система, совокупность и т.п.

Объекты, составляющие то или иное множество, называются его элементами.

Множество считается заданным, если для любого объекта можно установить, является он элементом данного множества или нет.

Множество иногда задается указанием всех его элементов. Если таких элементов немного, то для их перечисления используют обозначение: {x,y,z,…} - это множество, элементами которого являются объекты x, y, z и т.д. Порядок упоминания элементов при таком задании множества несуществен.

Более распространен способ задания множества путем указания свойства, которому должны удовлетворять все его элементы:
запись {x∈X∣∣xобладает свойствомP(x)}
выделяет те элементы множества X, которые удовлетворяют условию P(x).

Пусть A - произвольное множество, а a - какой-либо объект. Если a есть элемент A, то говорят, что a принадлежит A и пишут a∈A (читается: "a принадлежит A", "a - элемент A", "a из A"). Если же a не является элементом множества A, то это выражается следующим образом: a∉A (читается: "a не принадлежит A" или "a не есть элемент A").

Множество, не имеющее ни одного элемента, называется пустым и обозначается символом ∅. Например, множество всех прямоугольных равносторонних треугольников пусто, так как не существует ни одного такого треугольника.

Часто возникает такая ситуация, когда вводя множество математических объектов, удовлетворяющих определенному условию P(x), мы заранее не можем сказать, существует ли хотя бы один x, соответствующий этому условию P(x). Существование или несуществование таких x часто устанавливается лишь в конце исследования. До этого приходится говорить о множестве {x∣∣P(x)}, не зная заранее, будет это множество пустым или нет. В теории множеств пустое множество играет роль, аналогичную той, которая в арифметике принадлежит нулю.

Определение 1. Говорят, что A содержится в B или, иначе, A является подмножеством или частью B, если всякий элемент множества A является элементом множества B. При этом используют либо запись A⊂B, либо запись B⊃A (первая читается "A содержится в B", вторая - "B содержит A").

Свойства включения:

Для всякого множества A верно включение A⊂A.
Если A⊂B и B⊂C, то A⊂C.
Определение 2. Говорят, что множества A и B равны, и пишут A=B, если всякий элемент множества A принадлежит B и всякий элемент множества B принадлежит A.
Иначе говоря, A=B в том и только в том случае, если A⊂B и B⊂A.

Свойства равенства:

Для любого множества A имеет место равенство A=A.
Если A=B, то B=A.
Если A=B, B=C, то A=C.

**Подмножество**
Говорят, что A содержится в B или, иначе, A является подмножеством или частью B, если всякий элемент множества A является элементом множества B. При этом используют либо запись A⊂B (читается "A содержится в B").
Тем самым для множеств определено отношение, обозначаемое знаком ⊂, называется включением.

Составить: уравнения сторон $AB$, $AC$, $BC$

Решение:

Пусть $AA_1, BB_1, CC_1$ - медианы.

x-2y+1=0, 1-2*3+1=-4$\neq 0$

y-1=0, 3-1=2$\neq 0$

Следовательно, ни одна из данных медиан не выходит из A. Медианы пересекаются в одной точке.

$\begin{cases}
    x-2y+1=0,\\
   y-1=0;
 \end{cases}
 ⇔
 \begin{cases}
    x-2y+1=0,\\
   y=1;
 \end{cases}
  ⇔
 \begin{cases}
    x=1,\\
   y=1.
 \end{cases}
$

Назовём найденную точку точкой M.
Заметим, что абсцисса точки A = абсциссе точки M.
$|\vec{AM}|$ = $|y_M-y_A|$ = 2

Поскольку три медианы треугольника пересекаются в одной точке и делятся в ней в отношении 2:1, считая от вершины, то мы можем найти точку $A_1$ медианы $AA_1$, принадлежащую стороне BC.

$\vec{AA_1}=-3\vec{A_1M};\ A_1(\frac{1-3}{1-3}; \frac{3-3}{1-3}) ⇔ A_1(1; 0)$

$AA_1 : \frac{x-1}{1-1} = \frac{y-3}{0-3} \Leftrightarrow \frac{x-1}{0} = \frac{y-3}{-3} \Leftrightarrow x-1=0 \Leftrightarrow x=1 $

Пусть $l_1: x-2y+1=0,\ l_2: y-1=0,\ l_3: Ax+By+C=0$, где $l_1, l_2$ - медианы, $l_3$ - BC. Пусть $l_1$ - $BB_1$, $l_2$ - $CC_1$.

Поскольку $BA_1 =A_1C, AA_1$ - медиана, то $A_1(\frac{x_B+x_C}{2}; \frac{y_B+y_C}{2}) ⇒ \begin{cases}
    \frac{x_B+x_C}{2} = 1,\\
    \frac{y_B+y_C}{2} = 0;\\
 \end{cases}
 ⇔ \begin{cases}
    x_B+x_C = 2,\\
    y_B+y_C = 0;\\
    \end{cases}
 ⇔ \begin{cases}
    x_B = 2 - x_C,\\
    y_B = -y_C.\\
 \end{cases} $

 Получается, что $B=(x_B; y_B)=(2 - x_C; -y_C), \ C(x_C; y_C)$.

Тогда
$\begin{cases}
    l_3 \cap l_1 = B,\\
    l_3 \cap l_2 = C.\\
 \end{cases}$

То есть
$
 \begin{cases}
    Ax_B+By_B+C=0,\\
    x_B-2y_B+1=0,\\
    Ax_C+By_C+C=0,\\
    y_C-1=0;\\
 \end{cases}
⇔
 \begin{cases}
    A(2-x_C)-By_C+C=0,\\
    2 - x_C+2y_C+1=0,\\
    Ax_C+By_C+C=0,\\

 \end{cases}
 ⇔
 \begin{cases}
    A(2-x_C)-B+C=0,\\

 \end{cases}

 \begin{cases}
    -3A-B+C=0,\\
    x_C=5,\\
    5A+B+C=0,\\
    y_C=1.\\
 \end{cases}
$

Не будем продолжать дальше решение системы, поскольку мы уже получили коордианты вершин B и C: $C(5, 1), B=(2-5, -1) = B(-3, -1).$

Ответ: AB: -4x+4y-8=0, BC: 2x-8y-2=0, AC: -2x-4y+14=0.
